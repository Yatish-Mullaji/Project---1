# Data for 2006

```python
#import data from csv files
file_jobs_2006=pd.read_csv("Idea1/business_2006.csv")
file_jobs_2006.head()
```

```python
file_jobs_2006_clean=file_jobs_2006.rename(columns={'MSA DESCRIPTION': 'Metro Area',
                                                   'CODE': 'NAICS',
                                                   'NAICS DESCRIPTION': 'Industry',
                                                   'DATA TYPE': 'Filter',
                                                   'TOTAL': 'Total',
                                                   })
file_jobs_2006_clean.head()
```

```python
jobs_2006=file_jobs_2006_clean[['Metro Area', 'NAICS', 'Industry', 'Filter', 'Total']]
jobs_2006.head()
```

```python
jobs_2006['Industry'].unique()
```

```python
cities_2006=['San Francisco-Oakland-Fremont, CA Metropolitan Statistical Area',
       'Houston-Baytown-Sugar Land, TX Metropolitan Statistical Area',
       'Austin-Round Rock, TX Metropolitan Statistical Area',
       'Detroit-Warren-Livonia, MI Metropolitan Statistical Area',
       'Chicago-Naperville-Joliet, IL-IN-WI Metropolitan Statistical Area',
       'Boston-Cambridge-Quincy, MA-NH Metropolitan Statistical Area',
       'New York-Newark-Edison, NY-NJ-PA Metropolitan Statistical Area']

industries_2006=['Mining', 
            'Manufacturing', 
            'Information', 
            'Finance and Insurance', 
            'Professional, Scientific, and Technical Services']
            
filters_2006=['Establishments', 'Employment', 'Annual Payroll ($1,000)']
```

```python
jobs_2006_cities=jobs_2006[jobs_2006['Metro Area'].isin(cities_2006)]
jobs_2006_ind=jobs_2006_cities[jobs_2006_cities['Industry'].isin(industries_2006)]
jobs_2006_filtered=jobs_2006_ind[jobs_2006_ind['Filter'].isin(filters_2006)]
jobs_2006_filtered.to_csv('jobs_2006.csv')
jobs_2006_filtered.head()
```
